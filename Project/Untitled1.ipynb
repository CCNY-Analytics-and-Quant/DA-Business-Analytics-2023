```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create a figure and axis
fig, ax = plt.subplots()

# Plot the data
line, = ax.plot(x, y, color='blue', linewidth=2, linestyle='-')

# Set the edgecolor to black
line.set_edgecolor('black')

# Add labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Graph with Black Outline')

# Display the graph
plt.show()
```
