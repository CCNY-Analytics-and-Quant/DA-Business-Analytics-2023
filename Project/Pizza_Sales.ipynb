```python
import pandas as pd 

df = pd.read_csv('/Users/denissebeato/Documents/FinancialAnalysisCourse/JDL-SUM_23_PIZZA_PROJ/Project/SQL/JOIN_RESULTS.csv')
print(df)
```

```python
df['month'] = df['date'].str[5:7]
df['month'] = df['month'].astype('int')

print(df)
```

```python
import matplotlib.pyplot as plt
import calendar

results = df.groupby('month').sum()
months = range(1,13)
values = [5, 10, 8, 12, 6, 9, 3, 7, 8, 4, 11, 6]
plt.plot(months, values)

# Set the x-axis tick labels to abbreviated month names
abbreviated_month_names = [calendar.month_abbr[i] for i in months]
plt.xticks(months, abbreviated_month_names)


# Add labels and title
plt.title('Pizza Sales')

plt.bar(months, results['price'])
plt.ylim([60000, 77500])
plt.xticks(months)
plt.ylabel('Sales in USD')
plt.xlabel('MONTH')
```

```python
import matplotlib.pyplot as plt
import calendar

results = df.groupby('month').sum()
months = range(1, 13)

plt.plot(months, results['price'], linestyle='-', color='blue')

# abbreviated month names
abbreviated_month_names = [calendar.month_abbr[i] for i in months]
plt.xticks(months, abbreviated_month_names)

plt.grid(True)

plt.xlabel('Month')
plt.ylabel('Sales in USD')
plt.title('Pizza Sales')

plt.show()
```
