```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


top10_pizza = df['pizza_id'].value_counts().head(10).index.tolist()


df['Month'] = pd.to_datetime(df['date']).dt.month


filter_df = df[df['pizza_id'].isin(top10_pizza)]


group_df = filter_df.groupby(['pizza_id', 'Month'])['quantity'].sum().reset_index()


result = group_df.pivot(index='pizza_id', columns='Month', values='quantity')

month_abbreviations = {
    1: 'Jan',
    2: 'Feb',
    3: 'Mar',
    4: 'Apr',
    5: 'May',
    6: 'Jun',
    7: 'Jul',
    8: 'Aug',
    9: 'Sep',
    10: 'Oct',
    11: 'Nov',
    12: 'Dec'
}

result.columns = [month_abbreviations.get(month, '') for month in result.columns]

plt.figure(figsize=(10, 6))
ax = sns.heatmap(result, annot=True, fmt="g", cmap='Blues')


ax.set_ylabel('Top 10 Pizzas')
ax.set_xlabel('Months')

plt.show()
```
